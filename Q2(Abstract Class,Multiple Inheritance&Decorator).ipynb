### Demonstration of Multiple Inheritance

```python
class Father:
    def BG(self):
        print('Blood Group is A+')

class Mother:
    def color(self):
        print('Eye Color is brown')

class Child(Father,Mother):
    pass

c=Child()
print("Child's inherited qualities are:")
c.BG()
c.color()
```

### Demonstration of Abstract Class and Decorator

```python
import math
from abc import ABC, abstractmethod
class Myclass(ABC):
    @abstractmethod #use of decorator
    def calculate(self,x):
        pass 
#this is sub class 1 of Myclass
class Sub1(Myclass):
    def calculate(self,x):
        print('Square Value= ',x*x)

#this is sub-class 2 of Myclass
class Sub2(Myclass):
    def calculate(self,x):
        print('Square root= ',math.sqrt(x))
    
#this is sub-class 2 of Myclass
class Sub3(Myclass):
    def calculate(self,x):
        print('Cube value= ',x**3)

#creating Sub1 class object and calling calculate()
obj1=Sub1()
obj1.calculate(16)

#creating Sub2 class object and calling calculate()
obj2=Sub2()
obj2.calculate(16)

#creating Sub3 class object and calling calculate()
obj3=Sub3()
obj3.calculate(16)
```
